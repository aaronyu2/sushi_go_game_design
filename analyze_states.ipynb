```python
import pickle
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns

# disable scientific notation
pd.set_option('display.float_format', '{:.0f}'.format) 
# pd.set_option('display.precision', 0)
```

```python
table = []
with open("count_states_results", "r") as f:
    for line in f.readlines():
        trimmed = line.strip()[1:-1]
        table.append([int(num) for num in trimmed.split(", ")])
columns = [str(i + 1) for i in range(len(table[0]))]
num_states_df = pd.DataFrame(table, columns=columns)
num_states_df.head()
```

```python
start_states = sorted(num_states_df["1"].unique())
for start in start_states:
    start_df = num_states_df[num_states_df["1"] == start]
    if start_df.shape[0] > 5:
        print(f"STATES AFTER FIRST TURN: {start}")
        display(start_df.describe())
```

```python
sns.histplot(num_states_df["10"], bins=20)
plt.xlabel("Number of states")
plt.title("Number of ending states")
```

```python
plt.figure(figsize=(10, 6))
sns.histplot(num_states_df[num_states_df["6"] < 10000000]["6"], bins=20)
# sns.histplot(num_states_df["6"], bins=20)
plt.xlabel("Number of states (millions)")
plt.title("Number of states after turn 6")
plt.show()
```
